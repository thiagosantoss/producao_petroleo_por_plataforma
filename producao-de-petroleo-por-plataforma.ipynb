<h1>Regressão Linear Simples</h1>

<b><p>Iremos realizar uma regressão linear simples com o intuito de traçar uma reta que melhor representa os dados da média de produção de petróleo por periodo</p></b>


Formula da regressão (matricial):

$y = aX + b + \epsilon$

onde, 


$b=(X'X)^{-1}X'y$


Teoria:  [link](https://pt.wikipedia.org/wiki/M%C3%A9todo_dos_m%C3%ADnimos_quadrados)







```python
import numpy as np
```

<h4>Grafico da média de produção de petróleo por periodo</h4>

```python
grupo = df.groupby('Periodo')

y = np.array(grupo['Petroleo (bbl/dia)'].mean().values) # Pegando os dados
x = np.arange(0, 118) # gerando índices

fig, ax = plt.subplots(figsize=(15,7), dpi=500)

ax.set_xlabel('Período')
ax.set_ylabel('Petróleo (bbl/dia)')

plt.scatter(x, y) 
```

<h4>Aplicando a Regressão</h4>

```python
# Calculando
X = np.c_[np.ones(x.shape), x] 

b = np.linalg.pinv(X.T @ X) @ X.T @ y # Calculando b
y_hat = X @ b # Calculando o Y
```

```python
# Plotando o gráfico
fig, ax = plt.subplots(figsize=(15,7), dpi=500)

ax.set_xlabel('Período')
ax.set_ylabel('Petróleo')

plt.scatter(x, y) 
plt.plot(x, y_hat, color='green');
```

<b><p>Com isso, podemos afirmar que a produção média de Petróleo aumentou ao longo dos anos (de 2009 a 2018, como consta o Dataset) </p></b>
